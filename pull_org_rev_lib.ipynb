```python

new_column1_3 = ["broyage->tamisage", "tamisage->mélange", "mélange->extrusion", "extrusion->broyage_ug",
             "broyage_ug->tamisage_ug", "tamisage_ug->remplissage", "milieu->remplissage", 
              "remplissage->sortie_lyo", "sortie_lyo->retour_irr","lead_time"]

new_column6 = ["broyage->tamisage_(85:15)", "tamisage->mélange_(85:15)", "mélange->extrusion_(85:15)", 
               "extrusion->broyage_ug_(85:15)", "broyage_ug->tamisage_ug_(85:15)", "tamisage_ug->combinaison_(85:15)", 
               "broyage->tamisage_(75:25)", "tamisage->mélange_(75:25)", "mélange->extrusion_(75:25)", 
               "extrusion->broyage_ug_(75:25)", "broyage_ug->tamisage_ug_(75:25)", "tamisage_ug->combinaison_(75:25)", 
               "combinaison->remplissage",
               "milieu->remplissage", "remplissage->sortie_lyo", "sortie_lyo->retour_irr",
               "lead_time"]

#formulation 1 mois
for c in new_column1_3[::-1]:
    df_1.insert(loc = 6, column = c, value = None)
for c in new_column1_3[::-1]:
    df_3.insert(loc = 6, column = c, value = None)
for c in new_column6[::-1]:
    df_6.insert(loc = 6, column = c, value = None)
```

```python


df_1["DATE PREPA. MILIEU 1"] = pd.to_datetime(df_1["DATE PREPA. MILIEU 1"])
df_1["DATE SORTIE  LYO"] = pd.to_datetime(df_1["DATE SORTIE  LYO"])
df_1["DATE RETOUR IRRADIATION ET ECHANTILLONNAGE"] = pd.to_datetime(df_1["DATE RETOUR IRRADIATION ET ECHANTILLONNAGE"])

df_1.loc[:,"broyage->tamisage"] = df_1["3.75 MG DATE TAMISAGE POLYMERE"] - df_1["3.75 MG DATE Broyage Polymère"]
df_1.loc[:,"tamisage->mélange"] = df_1["3.75 MG DATE MELANGE "]- df_1["3.75 MG DATE TAMISAGE POLYMERE"]
df_1.loc[:,"mélange->extrusion"] = df_1["3.75 MG DATE EXTRUSION "] -df_1["3.75 MG DATE MELANGE "]
df_1.loc[:,"extrusion->broyage_ug"] = df_1["3.75 MG DATE Broyage BATONNETS "] - df_1["3.75 MG DATE EXTRUSION "]
df_1.loc[:,"broyage_ug->tamisage_ug"] = df_1["3.75 MG DATE TAMISAGE MICROGRANULES"] - df_1["3.75 MG DATE Broyage BATONNETS "]
df_1.loc[:,"tamisage_ug->remplissage"] = df_1["DATE DEBUT REMPLISSAGE "] - df_1["3.75 MG DATE TAMISAGE MICROGRANULES"]
df_1.loc[:,"milieu->remplissage"] = df_1["DATE DEBUT REMPLISSAGE "] - df_1["DATE PREPA. MILIEU 1"]
df_1.loc[:,"remplissage->sortie_lyo"] = df_1["DATE SORTIE  LYO"] - df_1["DATE DEBUT REMPLISSAGE "]
df_1.loc[:,"sortie_lyo->retour_irr"] = df_1["DATE RETOUR IRRADIATION ET ECHANTILLONNAGE"] - df_1["DATE SORTIE  LYO"]
df_1.loc[:,"lead_time"] = df_1["DATE RETOUR IRRADIATION ET ECHANTILLONNAGE"] - df_1["3.75 MG DATE Broyage Polymère"]
```
